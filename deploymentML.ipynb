```python
df['color'].unique()
```

THE HIERARCHY IS FROM D TO Z IN DESCENDING ORDER

DEF - highest

GHIJ

KLM

NOPQR

S to Z


```python
def enc_color(color):
    if color in 'DEF':
        return 5
    elif color in 'GHIJ':
        return 4
    elif color in 'KLM': 
        return 3
    elif color in 'NOPQR':
        return 2
    else: 
        return 1
```

```python
df.apply?
```

```python
df['color_ord'] = df['color'].apply(enc_color)
df
```

```python
X = df[['carat','cut_ord','clarity_ord','color_ord']]
y = df.price
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = train_test_split?
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 2)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```
